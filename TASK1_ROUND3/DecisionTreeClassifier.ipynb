```python
# We are using the OOPs Programming approach
# We will create a node for our Decision Tree model
# Each node in a decision tree can either be a decision node or a leaf node

class Node:
    def __init__(self,left=None,right=None,gain=None,value=None,feature_idx=None,threshold=None):

        # For a decision node
        self.feature_idx=feature_idx # stores the index of the feature we choose to split on
        self.threshold=threshold
        self.gain=gain # stores the value of information gain which ultimately determines the feature we choose to split on
        self.left=left # left child node of the parent node
        self.right=right # right child node of the parent node

        # For a leaf node it will only store the final value of the class it has converged to
        self.value=value
```

```python
# DecisionTreeClassifier class to find the best split and make predcition

class DecisionTree:
    def __init__(self,min_samples_split=2,max_depth=5):

        self.root=None

        self.max_depth=max_depth
        self.min_samples_split=min_samples_split


        self.feature_importances_=None

    # recursive function that takes the dataset as input for the current node
    def build_tree(self,dataset,current_depth=0):
        X,y=dataset[:,:-1],dataset[:,-1]

        n_samples,n_features=X.shape

        if n_samples >= self.min_samples_split and current_depth <= self.max_depth:
            best_split=self.best_split(dataset,n_features)

            if best_split["gain"] > 0 and len(best_split)>0:
                left_node=self.build_tree(best_split["left_dataset"],current_depth+1)
                right_node=self.build_tree(best_split["right_dataset"],current_depth+1)

                return Node(left=left_node,right=right_node,gain=best_split["gain"],feature_idx=best_split["feature_idx"],threshold=best_split["threshold"])
                        

        leaf_value=Counter(y).most_common(1)[0][0]
        return Node(value=leaf_value)

    def best_split(self,dataset,n_features):
        best_split={"feature_idx":None,
                    "threshold":None,
                    "gain":-float("inf"),
                    "left_dataset":None,
                    "right_dataset":None}

        for feature_idx in range(n_features):
            feature_values=dataset[:,feature_idx]
            unique_values=np.unique(feature_values)

            thresholds=(
                unique_values[:-1]+unique_values[1:]
            )/2


            for threshold in thresholds:
                left_dataset,right_dataset=self.split(dataset,feature_idx,threshold)

                if len(left_dataset) and len(right_dataset):
                    parent_y,left_y,right_y=dataset[:,-1],left_dataset[:,-1],right_dataset[:,-1]

                    gain=self.gini_gain(parent_y,left_y,right_y)

                    if gain > best_split["gain"]:
                        best_split["feature_idx"]=feature_idx
                        best_split["threshold"]=threshold
                        best_split["gain"]=gain
                        best_split["left_dataset"]=left_dataset
                        best_split["right_dataset"]=right_dataset

        return best_split


    def split(self,dataset,feature_idx,threshold):
        left_dataset=np.array([row for row in dataset if row[feature_idx] <= threshold])
        right_dataset=np.array([row for row in dataset if row[feature_idx] > threshold])

        return left_dataset,right_dataset


    def gini_gain(self,parent_y,left_y,right_y):
        left_weight=len(left_y)/len(parent_y)
        right_weight=len(right_y)/len(parent_y)


        gain=self.gini(parent_y) - (left_weight * self.gini(left_y) + right_weight * self.gini(right_y))

        return gain


    def gini(self,y):
        impurity=1

        class_labels=np.unique(y)
        for class_label in class_labels:
            p=np.sum(y==class_label)/len(y)
            impurity-=p**2

        return impurity


    def fit(self, X, y):
        dataset = np.concatenate((X, y.reshape(-1, 1)), axis=1)
        self.root = self.build_tree(dataset)

    def predict(self, X):    
        predictions = [self.predict_class(row, self.root) for row in X]
        return np.array(predictions)


    def predict_class(self,row,node):
        if node.value!=None:
            return node.value

        feature_value=row[node.feature_idx]
        if feature_value<=node.threshold:
            return self.predict_class(row,node.left)
        else:
            return self.predict_class(row,node.right)


    def print_tree(self,node=None,indent=" "):
        if node is None:
            node=self.root

        if node.value is not None:
            print(indent,"Leaf:",node.value)
            return

        print(indent,f"X[{node.feature_idx}] <= {node.threshold: .3f}")

        self.print_tree(node.left,indent + indent)

        self.print_tree(node.right,indent+indent)
```

```python
decision_tree_model=DecisionTree(min_samples_split=5,max_depth=5)
decision_tree_model.fit(X_train,y_train)

predictions=decision_tree_model.predict(X_test)

accuracy=accuracy_score(y_test,predictions)
print(f"Accuracy: {accuracy*100:.2f} %")
```

```python
decision_tree_classifier=DecisionTreeClassifier(
    criterion="gini",
    max_depth=5,
    random_state=42)

decision_tree_classifier.fit(X_train,y_train)

decision_tree_prediction=decision_tree_classifier.predict(X_test)


print("Sklearn model accuracy:",round(accuracy_score(y_test,decision_tree_prediction)*100,2),"%")

                                                     
```
